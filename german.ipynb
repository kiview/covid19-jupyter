```python
from urllib import request
import csv
import pandas as pd
import io
```

```python
url = "https://github.com/tsurai/covid-19-germany-csv/raw/master/states_time_series.csv"
req = request.Request(url)
response = request.urlopen(req).read().decode('utf8')
```

```python

series = pd.read_csv(io.StringIO(response))
series
```

```python
series.state.unique()
```

```python
nrw = series.loc[series['state'] == "Nordrhein-Westfalen"]
nrw
```

```python
nrw.cases.plot.bar(figsize=(10,10))
```
